```python
import numpy as np
import seaborn as sns
import matplotlib.pylab as plt
import math
```

```python
sp = (55, 37)
sc = 40
points = [
    (55.378 - sp[0], 37.120 - sp[1], 30 + 0.5 * 20),
    (55.809 - sp[0], 37.540 - sp[1], 20 + 0.5 * 10),
    (55.478 - sp[0], 37.760 - sp[1], 40 + 0.5 * 18)
]
```

```python
data = [np.zeros(shape=(sc,sc)),
        np.zeros(shape=(sc,sc)),
        np.zeros(shape=(sc,sc)),]

for (i, (x_target, y_target, S_target)) in enumerate(points):
    for (y,row) in enumerate(data[i]):
        iiy = y
        y /= sc
        for (x,_) in enumerate(row):
            iix = x
            x /= sc
            t_target = math.hypot(x_target-x, y_target-y)
            if t_target == 0:
                continue
            data[i][iiy,iix] = (S_target/t_target)/(sum(s1/math.hypot(x1-x, y1-y) for (y1,x1,s1) in points if math.hypot(x1-x, y1-y)!=0))
#for d in data:
    d /= sum( d2.sum() for d2 in data)
```

```python
ax = sns.heatmap(data[2], linewidth=0, cmap=sns.hls_palette(20, l=.3, s=.8))
plt.show()
```
